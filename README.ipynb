```julia
#implementação do B-multicast assíncrono.
function B_multicast_assync(g,m,B_deliver)
    for p in g
        remotecall(p, B_deliver, m)
    end
end

# Adding 3 processes (workers)
#addprocs(3)
addprocs(10)

# My group is composed by all workers
g = workers()

# Multicasting "oi!" message to group `g` by assuming `println` as B_deliver function.
for i = 1:50
    @time B_multicast_assync(g,"oi!",println)
end
```

Modifique a implementação do B-multicast para utilizar communicação síncrona.

```julia
#implementação do B-multicast síncrono.
function B_multicast_sync(g,m,B_deliver)
    for p in g
        remotecall_wait(p, B_deliver, m)
    end
end

# Adding 3 processes (workers)
#addprocs(3)
addprocs(10)

# My group is composed by all workers
g = workers()

# Multicasting "oi!" message to group `g` by assuming `println` as B_deliver function.
for i = 1:50
    @time B_multicast_sync(g,"oi!",println)
end
```

Execute as duas versões do B-multicast (síncrona e assíncrona) e compare as saídas.

A ordem das mensagens das saídas foram as mesmas para ambas versões do B-multicast? Por quê?

**Resposta**: _Não. Na versão síncrona o receptor sempre recebe as mensagens na ordem que foram enviadas pelo emissor, enquanto na versão assíncrona não existe mecanismos que garantam a ordem de chagada das mensagens no receptor. _

### Exercício 3: Análise de desempenho do R-multicast (3,0 pontos)

Implemente duas versões do algoritmo R-multicast: uma assíncrona e outra síncrona.

* R-multicast síncrono

```julia
#implementação do R-multicast síncrono.

received = Int[]

function R_multicast_sync(g, m, R_deliver)
    for p in g
        B_multicast_sync_check(received, p, g, m, R_deliver)
    end
end

function B_multicast_sync_check(received, p, g, m, R_deliver)
    if !(p in received)
        push!(received, p)
        remotecall_wait(p, R_deliver, m)
        R_multicast_sync(g, m, R_deliver)
    end
end

addprocs(10)

g = workers()

for i = 1:50
    @time R_multicast_sync(g, "Oi", println)
    received = Int[]
end
```

* R-multicast assíncrono

```julia
#implementação do R-multicast assíncrono.
#implementação do R-multicast síncrono.

received = Int[]

function R_multicast_assync(g, m, R_deliver)
    for p in g
        B_multicast_assync_check(received, p, g, m, R_deliver)
    end
end

function B_multicast_assync_check(received, p, g, m, R_deliver)
    if !(p in received)
        push!(received, p)
        remotecall(p, R_deliver, m)
        R_multicast_assync(g, m, R_deliver)
    end
end

addprocs(10)

g = workers()

for i = 1:50
    @time R_multicast_assync(g, "Oi", println)
    received = Int[]
end
```
